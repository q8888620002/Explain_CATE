```python
import argparse
import os
import sys
import pickle
import numpy as np
```

```python
cohort_name ="ist3"
shuffle = "True"

selection_type =[ 
    "if_pehe",
    "pseudo_outcome_r",
    "pseudo_outcome_dr"
]

learners = [
    "XLearner",
    "DRLearner",
    "SLearner",
    "TLearner",
    "RLearner",
    "RALearner",
    "TARNet",
    "DragonNet",
    "CFRNet_0.01",
    "CFRNet_0.001"
]

with open(f"results/{cohort_name}/model_selection_shuffle_{shuffle}.pkl", "rb") as output_file:
    results = pickle.load(output_file)
```

```python
for i in learners:
    print(i)
    print(f"qini score: {np.mean(results[i]['qini_score']):.4f} +- {2*np.std(results[i]['qini_score'])/np.sqrt(50):.4f}")
    for t in selection_type:
        print(f"{t}: {np.mean(results[i][t]):.4f} +- {2*np.std(results[i][t])/np.sqrt(20):4f}")

    print("\n")
```
