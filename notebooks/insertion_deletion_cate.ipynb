```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
file_path = "../results/ist3/insertion_deletion_xlearner_4.pkl"

results = pkl.load(open(file_path,'rb'))
results
```

```python
explainers = [
    "integrated_gradients", 
    "shapley_value_sampling", 
    # "naive_shap"
]


for exp_index, exp in enumerate(explainers):
    
    insertion_results = results[exp_index][2]
    deletion_results = results[exp_index][3]
        
    features_num = insertion_results.shape[0]
    plt.subplot(1,2,1)

    plt.plot([i for i in range(features_num)], insertion_results, label=exp)
    plt.title("insertion")

    plt.subplot(1,2,2)
    plt.plot([i for i in range(features_num)], deletion_results, label= exp)
    plt.title("deletion")
    
plt.legend(bbox_to_anchor=(1.5, 1))
```
